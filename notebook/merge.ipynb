```python
first_buy_df = trades_df.groupby(by='ticker').buy_date.min().reset_index()
first_buy_df
```

```python
trades_n_stats_df['total_cnt']  = 1
trades_n_stats_df['total_days'] = trades_n_stats_df.trading_days
trades_n_stats_df['mean_day']   = trades_n_stats_df.trading_days
trades_n_stats_df['daily_ret']  = trades_n_stats_df.daily_return
trades_n_stats_df
```
